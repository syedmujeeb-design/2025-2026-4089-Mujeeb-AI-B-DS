```python
# Creating a list
fruits = ["apple", "banana", "cherry"]

# Accessing elements
print(fruits[0])  # apple

# Adding elements
fruits.append("orange")

# Removing elements
fruits.remove("banana")

# Iterating through list
for fruit in fruits:
    print(fruit)

# Length of list
print(len(fruits))
```

```python
# Creating a tuple
colors = ("red", "green", "blue")

# Accessing elements
print(colors[1])  # green

# Tuples are immutable (cannot be changed)
# colors[0] = "yellow"  # This would cause an error

# Iterating through tuple
for color in colors:
    print(color)

# Length of tuple
print(len(colors))
```

```python
# Creating a dictionary
student = {
    "name": "John",
    "age": 20,
    "course": "Computer Science"
}

# Accessing values
print(student["name"])

# Adding/Updating values
student["age"] = 21
student["grade"] = "A"

# Removing a key-value pair
del student["course"]

# Iterating through dictionary
for key, value in student.items():
    print(key, ":", value)

# Length of dictionary
print(len(student))
```

```python
# Creating a set
numbers = {1, 2, 3, 4}

# Adding elements
numbers.add(5)

# Removing elements
numbers.remove(2)

# Sets do not allow duplicates
numbers.add(3)  # Duplicate, will not be added

# Iterating through set
for num in numbers:
    print(num)

# Length of set
print(len(numbers))
```
