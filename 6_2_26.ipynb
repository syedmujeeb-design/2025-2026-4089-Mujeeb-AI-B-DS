```python
# Creating sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

print(f"Set 1: {set1}")
print(f"Set 2: {set2}")

# Adding elements
set1.add(6)
print(f"Set 1 after adding 6: {set1}")

# Removing elements
set1.remove(1)
print(f"Set 1 after removing 1: {set1}")

# Union of sets
union_set = set1.union(set2)
print(f"Union of Set 1 and Set 2: {union_set}")

# Intersection of sets
intersection_set = set1.intersection(set2)
print(f"Intersection of Set 1 and Set 2: {intersection_set}")

# Difference of sets
difference_set = set1.difference(set2)
print(f"Difference (Set1 - Set2): {difference_set}")

# Symmetric Difference
symmetric_difference_set = set1.symmetric_difference(set2)
print(f"Symmetric Difference (elements in either set but not both): {symmetric_difference_set}")

# Check if a set is a subset of another
subset_check = {4, 5}.issubset(set1)
print(f"Is {{4, 5}} a subset of Set 1? {subset_check}")

# Check if a set is a superset of another
superset_check = set1.issuperset({4, 5})
print(f"Is Set 1 a superset of {{4, 5}}? {superset_check}")
```
