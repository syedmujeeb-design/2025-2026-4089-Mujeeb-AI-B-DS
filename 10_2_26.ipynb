```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [22, 25, 30, 28],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)

print("Full Data:")
print(df)
print("\n------------------\n")

print("Select Age column:")
print(df["Age"])
print("\n------------------\n")

print("Select Name and Marks columns:")
print(df[["Name", "Marks"]])
print("\n------------------\n")

print("Students with Marks > 80:")
print(df[df["Marks"] > 80])
print("\n------------------\n")
print("Names and Marks where Age > 23:")
print(df.loc[df["Age"] > 23, ["Name", "Marks"]])
```

```python
import pandas as pd

# Create sample data
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [22, 25, 30, 28, 21],
    "Marks": [85, 90, 78, 88, 92],
    "City": ["NY", "LA", "NY", "Chicago", "LA"]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)
print("\n--------------------\n")

filtered_marks = df[df["Marks"] > 85]
print("Marks > 85:")
print(filtered_marks)
print("\n--------------------\n")

filtered_age = df[df["Age"] >= 25]
print("Age >= 25:")
print(filtered_age)
print("\n--------------------\n")

filtered_multiple = df[(df["Marks"] > 85) & (df["City"] == "LA")]
print("Marks > 85 AND City = LA:")
print(filtered_multiple)
print("\n--------------------\n")

filtered_or = df[(df["City"] == "NY") | (df["Age"] < 23)]
print("City = NY OR Age < 23:")
print(filtered_or)
print("\n--------------------\n")
filtered_isin = df[df["City"].isin(["LA", "Chicago"])]
print("City in LA or Chicago:")
print(filtered_isin)
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

# Example
numbers = [64, 25, 12, 22, 11]
print(selection_sort(numbers))
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Example
numbers = [64, 34, 25, 12, 22]
print(bubble_sort(numbers))
```

```python
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i   # element found at index i
    return -1          # element not found

# Example
numbers = [10, 20, 30, 40, 50]
key = 30

result = linear_search(numbers, key)

if result != -1:
    print("Element found at index", result)
else:
    print("Element not found")
```

```python
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1

    return -1

# Example (sorted list)
numbers = [10, 20, 30, 40, 50]
key = 40

result = binary_search(numbers, key)

if result != -1:
    print("Element found at index", result)
else:
    print("Element not found")
```

```python
import pandas as pd

# Sample data
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [22, 25, 30, 28, 21],
    "Marks": [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)
print("Original Data:\n", df, "\n")

# 1. Sort by Marks descending
df_sorted = df.sort_values(by="Marks", ascending=False)
print("Sorted by Marks (descending):\n", df_sorted, "\n")

# 2. Filter students with Marks > 80
df_filtered = df_sorted[df_sorted["Marks"] > 80]
print("Filtered (Marks > 80):\n", df_filtered, "\n")

# 3. Rearrange columns
df_rearranged = df_filtered[["Name", "Marks", "Age"]]
print("Columns Reordered:\n", df_rearranged, "\n")

# 4. Reset index
df_final = df_rearranged.reset_index(drop=True)
print("Final Data (Index Reset):\n", df_final)
```

```python
import pandas as pd

# Sample data
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Marks": [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)
print("Original Data:\n", df, "\n")

# Rank by Marks (highest marks get rank 1)
df["Rank"] = df["Marks"].rank(ascending=False, method='dense')

# Sort by Rank
df_sorted = df.sort_values("Rank")
print("Data with Ranks:\n", df_sorted.reset_index(drop=True))
```
