```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Iris dataset (no headers in file)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Read dataset and manually assign column names
df = pd.read_csv(url, header=None)
df.columns = ["sepal_length", "sepal_width",
              "petal_length", "petal_width", "species"]

print("Dataset (first 10 rows):")
print(df.head(10))

# Select column
data = df["sepal_length"]

print("\nSummary Statistics:")
print(data.describe())

print("\nVariance:", np.var(data))
print("Standard Deviation:", np.std(data))

# Plot histogram
plt.hist(data, bins=10, color='skyblue', edgecolor='black')

# Mean line
plt.axvline(np.mean(data),
            color='red',
            linestyle='dashed',
            linewidth=2,
            label=f"Mean: {np.mean(data):.2f}")

# Median line
plt.axvline(np.median(data),
            color='green',
            linestyle='dashed',
            linewidth=2,
            label=f"Median: {np.median(data):.2f}")

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.legend()
plt.show()

# Inference
mean_value = np.mean(data)
median_value = np.median(data)

print("\nInference:")
print("Mean value:", mean_value)
print("Median value:", median_value)

if mean_value > median_value:
    print("Data is slightly right-skewed")
elif mean_value < median_value:
    print("Data is slightly left-skewed")
else:
    print("Data is symmetric")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data
data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]

# Create DataFrame
df = pd.DataFrame(data, columns=["Values"])

print("Dataset:")
print(df)

print("\nSummary Statistics:")
print(df.describe())

print("\nVariance:", np.var(data))
print("Standard Deviation:", np.std(data))

# Histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Data Distribution")
plt.show()

# Mean and Median
mean_value = np.mean(data)
median_value = np.median(data)

print("\nInference")
print("Mean Value:", mean_value)
print("Median Value:", median_value)

# Skewness check
if mean_value > median_value:
    print("Data is slightly right-skewed")
elif mean_value < median_value:
    print("Data is slightly left-skewed")
else:
    print("Data is symmetric")
```
